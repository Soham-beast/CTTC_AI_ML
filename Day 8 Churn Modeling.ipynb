```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\CTTC1\Desktop\Churn_Modelling.csv")
data
```

```python
#distribution plot

sns.distplot(data.CreditScore[data.Exited==1],hist=False)
sns.distplot(data.CreditScore[data.Exited==0],hist=False)
plt.legend(['Leaving','Not Leaving'])
plt.show()
```
