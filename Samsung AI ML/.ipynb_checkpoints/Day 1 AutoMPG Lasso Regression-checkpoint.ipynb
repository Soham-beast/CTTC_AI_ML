# Package Import

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Wrangling/Reading

```python
data = pd.read_csv(r"C:\Users\CTTC\Downloads\Datasets\auto-mpg.data",header=None,delim_whitespace=True)   #make the whitespaces as column separator
data
```

```python
data.columns = ['mpg','cyclinders','displacement','horsepower','weight','accelaration','model_year','origin','car_name']     #column renaming
data
#renaming the columns with the dataset attributes
```

# Data Cleaning

```python
data.isnull().sum()    #total number of NULL elements in each column
```

```python
data.describe(include='all')        #describes about all the numeric and string valued columns
```

```python
for i in data.columns:
    print(i,':',sum(data[i]=='?'))           #checks if the missing data denoted by '?' is present or not 
```

```python
d1 = data.describe(include='all')        #describes both the numeric and string columns and stores them in 'd1'
```

```python
data['horsepower'].replace('?',d1['horsepower'][2],inplace=True)       #replacing '?' with the top repeated value
```

```python
data
```

```python
for i in data.columns:
    print(i,':',sum(data[i]=='?'))           #checks if the missing data denoted by '?' is present or not 
```

```python
data.dtypes             #checks the data types of each column
```

```python
data['horsepower']=data['horsepower'].astype('float64')     #changing the datatype of the column to 'float64'
```

```python
data.dtypes             #checks the data types of each column
```

```python
data.drop(['car_name'],axis=1,inplace=True)          #removing unwanted columns that are not needed for our ML model
data

#we mention 'inplace' as 'True' for making changes in the original sheet also
```

# Ip/Op Creation

```python
ip = data.drop('mpg',axis=1)        #all data column except 'mpg'
op = data.mpg                       #only the 'mpg' data
```

# Train Test Split

```python
#sklearn is ML library, scikit learn

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ip,op,test_size=0.2,random_state=2)
x_train
#x is input data and y is output data, divided into train and| test sets, total 4 parts
#test size of 0.2 denotes that 20% data is for test and 80% data is for training the model
#the splitted data is selected at random from the dataset, as by default 'random_state' is 'none'
#if you do not want a random selection, we can define 'random_state' with a specific value
#if the 'random_state' is 5, then the data taken during the 5th time is considered as constant
```

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
#brings all the data entries into one range, using the data from 'z', so that the data do not deviate from each other
#as the data is in a range, the calculation becomes easy and data is continous
```

# ML Model Training

```python
#'lasso' automatically takes up the important features by itself, and reduce overfitting
#it is also called 'l1 regression'
#'lasso' stands for Least Absolute Shrinkage and Selection Operator
#the coefficient count is reduced and hence reduce model complexity
#here also 'regularization parameter' and 'penalty factor' are taken into consideration

from sklearn.linear_model import Lasso
L = Lasso(alpha=0.01)
L.fit(x_train,y_train)
```

# Prediction

```python
yp = L.predict(x_test)          #'yp' means y-predicted; y was the test sample already stored
```

```python
yp
```

# Accuracy Check

```python
#determines the error by squaring the displacement between the predicted data and the actual data
#the sum of squared errors is the cost function of the dataset
#the Sum of Squared Errors(SSE) is the Mean Squared Errors(MSE) multiplied by the degrees of freedom for errors for the constrained model, which is n-2(k+1)

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

b = mean_squared_error(y_test,yp)
c = r2_score(y_test,yp)

print('Mean Squared Error is:',b)
print('R2 Score is:',c)
```
