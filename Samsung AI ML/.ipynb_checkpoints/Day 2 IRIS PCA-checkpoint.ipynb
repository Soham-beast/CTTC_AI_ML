```python
#'PCA'-> Principal Component Analysis
#a popular unsupervised learning technique for dimensionality reduction of data, i.e. number of input columns are reduced
#for 2-D plot we can reduce the dataset to 2 input columns and for 3-D plots it is 3.
#for dimensional reduction, the mean of each feature is found then the covariance is calculated
#the covariance matrix is formed for each value of covariance pairs
#the eigen values and eigen vectors are calculated from: 

#It increases interpretability yet, at the same time, it minimizes information loss. 
#It helps to find the most significant features in a dataset and makes the data easy for plotting in 2D and 3D.
#It is a tool for identifying the main axes of variance within a data set and allows for easy data exploration to understand the key variables in the data and spot outliers.
#PCA is an unsupervised learning algorithm while LDA is a supervised learning algorithm. 
#This means that PCA finds directions of maximum variance regardless of class labels while LDA finds directions of maximum class separability.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\CTTC\Downloads\Datasets\iris.data",header=None)
data
```

```python
data.columns = ['SL','SW','PL','PW','flower']   
data
```

```python
for i in data.columns:
    print(i,'\n',data[i].unique(),'\n')
```

```python
data.dtypes
```

```python
#Label Encoding

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['flower'] = le.fit_transform(data['flower'])
data['flower'].unique()
```

```python
#ip/op creation

ip=data.drop('flower',axis=1)
op=data.flower
```

```python
data.head()
```

```python
#Data Standardization

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

#standardizing the input data
ip = sc.fit_transform(ip)
```

```python
ip
```

```python
#PCA

from sklearn.decomposition import PCA
pc = PCA(n_components=2)
#'n_components' reduces the dimensions to 2
p_comp = pc.fit_transform(ip)
```

```python
p_comp.shape

#there are still 150 rows, but the 4 input columns are reduced to 2 only
```

```python
p_df = pd.DataFrame(data=p_comp,columns=['Principal Component 1','Principal Component 2'])
p_df
```

```python
df_final = pd.concat([p_df,op],axis=1)
df_final
```

```python
plt.figure(figsize=(15,10))
plt.scatter(df_final['Principal Component 1'],df_final['Principal Component 2'],c=op,cmap='plasma')
plt.show()
```
