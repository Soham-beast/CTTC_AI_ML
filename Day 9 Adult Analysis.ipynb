```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\CTTC1\Downloads\adult.data",header=None,delimiter=' *, *',engine='python')
data     
#'header' makes the first row to come to the dataset adn not get transfered to the column heading
#'delimiter' removes the blank spaces from front and back of each input data
#'engine' defines the kernel engine as 'python'
```

```python
data.columns = ['age','work_class','fnlwgt','education','education_num','maritial_status','occupation','relationship','race','sex','capital_gain','capital_loss','hours_per-week','native_country','Salary']
data
#renaming the header as not as indexes, but as attributes
```

```python
for i in data.columns:
    print(i,':',sum(data[i]=='?'))
#checks the number of '?' in each column using the 'sum' command
```

```python
d1 = data.describe(include='all')
d1
#stores the description in terms of both numeric and string valued columns in another data set variable 
```

```python
for i in ['work_class','occupation','native_country']:
    data[i].replace('?',d1[i][2],inplace=True)
#Replaces the missing datas having '?' with the top repeated values in that columns
```

```python
for i in data.columns:
    print(i,':',sum(data[i]=='?'))
#checks the number of '?' in each column using the 'sum' command
```

# Histogram age<->salary

```python
sns.displot(data.age[data.Salary=='>50K'])     #shows the histogram 
sns.displot(data.age[data.Salary=='<=50K'])    #shows the histogram
plt.legend(['More salary','Less salary'])
plt.show()
```

# KDE age<->salary

```python
sns.kdeplot(data.age[data.Salary=='>50K'])     #shows the kernel density estimation 
sns.kdeplot(data.age[data.Salary=='<=50K'])    #shows the kernel density estimation
plt.legend(['More salary','Less salary'])
plt.show()
```
